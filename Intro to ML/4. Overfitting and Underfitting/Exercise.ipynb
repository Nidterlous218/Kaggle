```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_absolute_error
def get_mae(max_leaf_nodes, train_X, val_X, train_y, val_y):
    model = DecisionTreeRegressor(max_leaf_nodes = max_leaf_nodes, random_state = 0)
    model.fit(train_X, train_y)
    preds_val = model.predict(val_X)
    mae = mean_absolute_error(preds_val, val_y)
    return(mae)
    
```

```python
import pandas as pd
```

```python
home_data = pd.read_csv('train.csv')
home_data.columns
```

```python
y = home_data['SalePrice']
features = ['LotArea', 'YearBuilt', '1stFlrSF', '2ndFlrSF', 'FullBath', 
            'BedroomAbvGr', 'TotRmsAbvGrd']
X = home_data[features]
```

```python
from sklearn.model_selection import train_test_split
train_X, val_X, train_y, val_y = train_test_split(X, y, random_state = 0)
```

```python
iowa_model = DecisionTreeRegressor(random_state = 1)
iowa_model.fit(train_X, train_y)

val_predictions = iowa_model.predict(val_X)
val_mae = mean_absolute_error(val_predictions, val_y)
print("Validation MAE: " + str(int(val_mae)))
```

```python
candidate_max_leaf_nodes = [5, 25, 50, 100 ,250, 500]
for max_leaf_nodes in candidate_max_leaf_nodes:
    results = get_mae(max_leaf_nodes, train_X, val_X, train_y, val_y)
    print("Max_Leaf_Nodes: " + str(max_leaf_nodes) + "\t\t" + "MAE: " + str(results))
```
